```python
def maxelement(l):
    dl=dict()
    for i in l:
        dl[i]=dl.get(i,0)+1
    ldl=sorted(dl, key=lambda x:dl[x], reverse=True)
    if len(ldl)>1:
        if ldl[0]=="O":
            tag=ldl[1]
        else:
            tag=ldl[0]
    else:
        tag=ldl[0]
        
    return tag
```

```python
#creating a list of index of heads for a given word
def heads_chain(n,lhead):
  lh=list()
  #for i in 
  #if "0" not in lhead:
  #  print("dependency tree error")
  c=0
  while n!= int(lhead[n-1]):
    n=int(lhead[n-1])
    #n=int(n)
    #print(n)
    lh.append(n)
    c+=1
    if c>= len(lhead):
      break
  return lh
```

```python
#creating a list of index of dependents for a given word
def dependent_chain(n,lhead):
  ld=list()
  for i in range(len(lhead)):
    if n in heads_chain(i+1,lhead):
      ld.append(i+1)
  return ld
```

```python
def is_superior(n1,n2, lhead):
  if n1 in heads_chain(n2,lhead):
    return True
  else:
    return False

```

```python
def max_indexer(recorder,lhead):
  m=recorder[0]
  for i in recorder:
    i1=m[0]
    i2=i[0]
    if is_superior(i2,i1,lhead):
      m=i
  return m
```

```python
def uncode_events(file1, file2,):
  lle=list()
  lld=dict()
  with open(file1, "r", encoding="utf-8") as e:
    for l in e.readlines():
      le=l.strip().split("||")
      lle.append(le)
    with open(file2, "r", encoding="utf-8") as d:
      for l in d.readlines():
        if l[0]!="#":
          ld=l.split("\t")
          if len(ld)>2:
            lld[ld[0]]=ld[1]
  nll=list()
  for l in lle:
    nl=list()
    for e in l[:-3]:
      if e != "Non":
        re=lld[e]
      else:
        re=""
      #nl.append(re)
      nl.append(re.replace("_",' '))
    nl+=l[-3:]
    #nll.append(nl)
    nll.append(nl)
  return nll
```

```python
def date2table(date,indice):
  ld=date.split(" ")
  nld=ld.copy()
  if len(nld) > 3:
    if indice=="debut":
      nld=nld[:3]
    else:
      nld=nld[-3:]
  nld.reverse()
  
  nnld=list()
  while len(nld[0])<4:
    nld.pop(0)
    if len(nld)<1:
      break
  while len(nld) < 4 :
    nld.append("")
  #if len(nld) >4:
  #  print(nld)
  return "\t".join(nld)+"\t"
```
